```python

class Agente:

    def __init__(self, nome="inimigo", pred=1):
        self.state = "idle"
        self.nome = nome
        self.pred = pred

    def update_state(self, clicked, success, vida):

        if self.state == "idle":

            if not vida:
                self.state = "morrer"

            elif clicked and success:
                self.state = "elogiar"

            elif clicked and not success:
                self.state = "investigar"


        elif self.state == "investigar":

            if not vida:
                self.state = "morrer"

            elif success:
                self.state = "atacar"

            else:
                self.state = "procurar"


        elif self.state == "procurar":

            if success:
                self.state = "atacar"
            else:
                self.state = "idle"


        elif self.state == "atacar":

            if success:
                self.state = "elogiar"
            else:
                self.state = "invocar"


        elif self.state == "invocar":

            if not vida:
                self.state = "morrer"
            else:
                self.state = "idle"


        elif self.state == "elogiar":

            if not vida:
                self.state = "morrer"
            else:
                self.state = "idle"


    def perform_action(self):

        print("Estado atual:", self.state)

        if self.state == "elogiar":
            print("NPC elogia o jogador")

        elif self.state == "invocar":
            print("NPC invoca reforços")

        elif self.state == "investigar":
            print("NPC está investigando o jogador")

        elif self.state == "procurar":
            print("NPC está procurando o jogador")

        elif self.state == "atacar":
            print("NPC ataca o jogador")

        elif self.state == "morrer":
            print("NPC morreu")
            return True

        return False


# ---------------------------
# Simulação do jogo
# ---------------------------

agente_vidas = [True, True, True, True, True, False]

player_clikeds = [False, True, True, True, True, False]

player_success = [False, False, True, False, True, False]


agente = Agente()


for vida, clicked, success in zip(agente_vidas, player_clikeds, player_success):

    agente.update_state(clicked, success, vida)

    acabou = agente.perform_action()

    if acabou:
        break
```

```python
import random


class Agente:

    def __init__(self, nome="npc"):
        self.state = "idle"
        self.nome = nome


    def update_state(self, success):

        if self.state == "idle":

            print("##################")

            numero = random.randint(0,10)
            print("Número sorteado:", numero)

            if numero % 2 == 0:
                self.state = "elogiar"
            else:
                self.state = "investigar"


        elif self.state == "elogiar":

            self.state = "idle"


        elif self.state == "investigar":

            self.state = "procurar"


        elif self.state == "procurar":

            self.state = "atacar"


        elif self.state == "atacar":

            if success:
                self.state = "morrer"
            else:
                self.state = "invocar"


        elif self.state == "invocar":

            self.state = "morrer"


    def perform_action(self):

        print("Estado atual:", self.state)

        if self.state == "elogiar":
            print("NPC elogia o jogador")

        elif self.state == "investigar":
            print("NPC iniciou investigação")

        elif self.state == "procurar":
            print("NPC está procurando o jogador")

        elif self.state == "atacar":
            print("NPC está atacando")

        elif self.state == "invocar":
            print("NPC invoca reforços")

        elif self.state == "morrer":
            print("Jogador morreu")
            return True

        return False



# -------------------------
# Simulação contínua
# -------------------------

agente = Agente()

jogador_vivo = True

while jogador_vivo:

    # evento de sucesso aleatório do jogador
    success = random.choice([True, False])

    agente.update_state(success)

    jogador_morreu = agente.perform_action()

    if jogador_morreu:
        jogador_vivo = False
```
